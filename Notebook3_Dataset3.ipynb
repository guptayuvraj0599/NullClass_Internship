```python
import pandas as pd
import random
from faker import Faker
import datetime

# Initialize Faker
fake = Faker()

# Function to generate random job titles
def get_job_title():
    return random.choice(['Data Scientist', 'Art Teacher', 'AeroSpace Engineer'])

# Function to generate random salary (greater than $10k)
def get_salary():
    return round(random.uniform(10000, 150000), 2)

# Function to generate random experience (greater than 2 years)
def get_experience():
    return random.randint(3, 20)

# Function to generate random job posting date (before 08/01/2023)
def get_job_posting_date():
    start_date = datetime.date(2020, 1, 1)
    end_date = datetime.date(2023, 7, 31)
    return fake.date_between(start_date=start_date, end_date=end_date)

# Function to generate random posting time (between 12 PM and 6 PM)
def get_posting_time():
    # Generate time within the range 12:00 PM to 6:00 PM
    hour = random.randint(12, 17)  # 12 PM to 5 PM
    minute = random.randint(0, 59)
    return datetime.time(hour, minute).strftime('%H:%M')

# Function to generate random company size (greater than 80,000)
def get_company_size():
    return random.randint(80001, 500000)

# Function to get random latitude and longitude for a given country
def get_location(country):
    if country == 'India':
        # Approximate range for India
        latitude = random.uniform(8.4, 37.6)
        longitude = random.uniform(68.7, 97.25)
    elif country == 'Germany':
        # Approximate range for Germany
        latitude = random.uniform(47.2, 54.9)
        longitude = random.uniform(5.87, 15.04)
    return latitude, longitude

# Generate a large dataset
def generate_dataset(num_records):
    data = []
    for _ in range(num_records):
        country = random.choice(['India', 'Germany'])
        latitude, longitude = get_location(country)
        record = {
            'Country': country,
            'Qualification': 'B.Tech',
            'Work Type': 'Full time',
            'Experience (Years)': get_experience(),
            'Job Title': get_job_title(),
            'Salary ($)': get_salary(),
            'Job Portal': 'Indeed',
            'Gender Preference': 'Female',
            'Job Posting Date': get_job_posting_date(),
            'Job Posting Time': get_posting_time(),
            'Company Size': get_company_size(),
            'Latitude': latitude,
            'Longitude': longitude
        }
        data.append(record)

    # Create a DataFrame
    df = pd.DataFrame(data)
    return df

# Generate 1000 records (adjust the number as needed)
df = generate_dataset(1000)

# Save the dataset to a CSV file
df.to_csv('Data3_Task3.csv', index=False)
```
